```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
train = pd.read_csv(r'C:\Users\rakhe\OneDrive\Desktop\ElevateLabs\Task5\titanic\train.csv')
test = pd.read_csv(r'C:\Users\rakhe\OneDrive\Desktop\ElevateLabs\Task5\titanic\test.csv')
gender = pd.read_csv(r'C:\Users\rakhe\OneDrive\Desktop\ElevateLabs\Task5\titanic\gender_submission.csv')
train.info()
test.info()
gender.info()

```

```python
train.describe()
test.describe()
```

```python
train['Sex'].value_counts()
train['Pclass'].value_counts()
train['Embarked'].value_counts()
```

```python
train.isnull().sum()
test.isnull().sum()
```

```python
sns.heatmap(train.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Missing Data Heatmap - Train Set')
plt.show()
```

```python
Observation:

Fare has only 1 missing values.
Age has 86 missing values.
Cabin has 327 missing values.
```

```python
train['Age'].hist(bins=30, edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
Observations:
Most of the passengers are between ages 20-40.
Few children and elderly passengers.
```

```python
sns.boxplot(x='Pclass', y='Fare', data=train)
plt.title('Fare Distribution by Pclass')
plt.show()
```

```python
Observation:

Fare tends to be higher in 1st class.
3rd class fares are generally lower, with few exceptions.
```

```python
sns.countplot(x='Survived', data=train)
plt.title('Survival Counts')
plt.show()
```

```python
Observations:
The number of people who died  is greater than the number of people who lived (survived).
```

```python
sns.barplot(x='Sex', y='Survived', data=train)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
Observation:

Females had a much higher survival rate than males.
```

```python
sns.barplot(x='Pclass', y='Survived', data=train)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
Observation:

1st class passengers had the highest survival rate, followed by 2nd and then 3rd class.
```

```python
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=train)
plt.title('Age vs Fare by Survival Status')
plt.show()
```

```python
Observations:

People who survived are mostly found within specific groups of fare amounts and ages.
Passengers who paid higher ticket prices seem to have a better chance of surviving.
```

```python
sns.pairplot(train[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
plt.show()
```

```python
Observation:

There are clear separations in survival for different classes and fares.
Age shows some overlap but still varies between survival groups.
```
